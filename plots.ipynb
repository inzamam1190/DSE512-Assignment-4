```python
import matplotlib as mpl
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
p1 = pd.read_csv('metrics_1.csv')
p2 = pd.read_csv('metrics_2.csv')
p4 = pd.read_csv('metrics_4.csv')
p8 = pd.read_csv('metrics_8.csv')
```

```python
p4 = p4.iloc[:30,:]
p8 = p8.iloc[:30,:]
```

```python
mpl.rcParams['figure.figsize'] = [8,6]
mpl.rcParams['figure.dpi'] = 200
```

```python
plt.plot(p1.epoch_accuracy, label='p=1')
plt.plot(p2.epoch_accuracy, label='p=2')
plt.plot(p4.epoch_accuracy, label='p=4')
plt.plot(p8.epoch_accuracy, label='p=8')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Training Accuracy')
plt.title('Epoch accuracy for different number of processors')
```

```python
plt.plot(p1.elapsed_time, label='p=1')
plt.plot(p2.elapsed_time, label='p=2')
plt.plot(p4.elapsed_time, label='p=4')
plt.plot(p8.elapsed_time, label='p=8')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Elapsed Time')
plt.title('Elapsed time in training for different number of processors')
```

## Summary

As we can see from the second plot, when we increase the number of processes available, training time gets smaller hence we achieve speedup. This is exactly what I expected and it's the ideal scenario. Gustafson's scalability law is the most applicable in this case. The effective batch size increases with the number of processes, but as we can see from the first plot, the accuracies doesn't change that much, approximately same at 80%. Again, this is the ideal case and exactly what I expected.
